```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/wpbc.csv")
df['Lymph Node Status'] = df['Lymph Node Status'].fillna(int(df['Lymph Node Status'].mean()))
df = (df - df.mean())/df.std()
```

```python
plt.scatter(df["Tumor Size"], df["Lymph Node Status"])
plt.xlabel("Tumor Size (Normalized)")
plt.xlim(-2,4)
plt.ylabel("Lymph Node Status (Normalized)")
plt.ylim(-1,3)
plt.grid(True)
```

```python
x1 = np.array(df['Tumor Size'])
x2 = np.array(df['Lymph Node Status'])
A = np.array([(x1[i], x2[i]) for i in range(len(x1))])

def cluster_assignment(data, w, b):
    
    m, n = data.shape
    k = w.shape[0]

    distance_matrix = np.abs((data @ np.transpose(w))-b)
    assignment = np.zeros(m)
    
    for i in range(m):
        assignment[i] = np.argmin(distance_matrix[i]) + 1
        
    return assignment

def cluster_update(data, w, b, assignment):
    
    m, n = data.shape
    k = w.shape[0]
    
    for cluster in range(k):
        A_cluster = data[assignment == cluster+1]
        m_cluster = len(A_cluster)
        
        e = np.ones((m_cluster, 1))
        eT = np.transpose(e)
        centering_matrix = np.identity(m_cluster) - ((e @ eT)/m_cluster)
        
        B = np.transpose(A_cluster) @ centering_matrix @ A_cluster
        
        eigenval, eigenvec = np.linalg.eig(B)
        min_eigenval_index = np.argmin(eigenval)
        smallest_eigenvec = eigenvec[:, min_eigenval_index]
        
        w[cluster] = smallest_eigenvec
        b[cluster] = (eT @ A_cluster @ w[cluster])/m_cluster

def get_within_cluster_sum_of_squares(data, w, b, assignment):
    m, n = data.shape
    k = w.shape[0]
    
    distance = 0
    distance_matrix = np.abs((data @ np.transpose(w))-b)
    
    for i in range(m):
        distance += pow((distance_matrix[i][int(assignment[i])-1]), 2)

    return distance

def plot_clusters_and_cluster_plane(data, w, b, assignment):
    m,n = data.shape
    k = w.shape[0]

    plt.figure(figsize=(12,8))
    colors = ["red", "blue", "yellow"]
    for cluster in range(k):
        cluster_data = data[assignment == cluster+1]
        plt.scatter(cluster_data[:, 0], cluster_data[:, 1], c=colors[cluster], label=f'Cluster {cluster + 1}')
        weight = w[cluster]
        bias = b[cluster]
        x_vals = np.linspace(-3, 5, 500)
        y_vals = (-weight[0] * x_vals - bias) / weight[1]
        plt.plot(x_vals, y_vals, c=colors[cluster], label=f'Cluster Plane {cluster + 1}', linewidth=2)
    
    plt.xlabel('X1')
    plt.xlim((-2,4))
    plt.ylabel('X2')
    plt.ylim((-1,3))
    plt.title('k-Plane Clustering')
    plt.legend()
    plt.grid(True)
    
def kPlaneClustering(data, n_clusters, n_iterations, random_seed=None):
    
    if random_seed is not None:
        np.random.seed(random_seed)

    m, n = data.shape
    k = n_clusters

    w = np.random.rand(k,n)
    w /= np.linalg.norm(w, axis=1, keepdims=True)
    b = np.random.rand(k)
    
    old_assignment = np.random.choice([1,2,3], size= m)
        
    for it in range(n_iterations):
        new_assignment = cluster_assignment(data, w, b)
        cluster_update(data, w, b, new_assignment)

        if (np.array_equal(old_assignment, new_assignment)):
            print(f"Algorithm converged before specified iterations!\nIterations taken -> {it+1}")
            break
        old_assignment = new_assignment.copy()

    return new_assignment, w, b
    
final_assignment, final_weights, final_bias = kPlaneClustering(A,3,50,42)
plot_clusters_and_cluster_plane(A,final_weights, final_bias, final_assignment)
```

```python
num_clusters = list(range(1,8))
sum_of_squares = []

for clusters in num_clusters:
    final_assignment, final_weights, final_bias = kPlaneClustering(A,clusters,20,42)
    distance = get_within_cluster_sum_of_squares(data, final_weights, final_bias, final_assignment)
    sum_of_squares.append(distance)

plt.figure(1)
plt.plot(num_clusters, sum_of_squares)
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
```
